```python
import pygrib
import numpy as np
```

```python
d_path=r"D:\Tommy\++AI\++datasets\weather_data\gfs.t18z.pgrb2.0p25.f000"
grbs = pygrib.open(d_path)
temp100 = grbs[642]
Ugrd100 = grbs[643]
Vgrd100 = grbs[644]
```

```python
temp = temp100.values - 273.15
lats, lons = temp100.latlons()
print(temp)
print(lats)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data (replace with your actual data)
# In this example, we have a 2D array of temperature values for each location (latitude and longitude)
latitudes = lats
longitudes = lons
temperatures = temp

# Create the heat map
fig, ax = plt.subplots(figsize=(20,12))
heatmap = ax.pcolormesh(longitudes, latitudes, temperatures, cmap='jet',shading='auto')

# Add a color bar legend
cbar = fig.colorbar(heatmap, ax=ax)
cbar.set_label('Temperature (°C)')

# Set axis labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Weather Temperature Heat Map')

plt.show()
# Save the heat map to an image file
#plt.savefig('temperature_heatmap.png')
```
